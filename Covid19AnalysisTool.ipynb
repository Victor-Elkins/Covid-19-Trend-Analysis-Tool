```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

df = pd.read_csv('us-states.csv') #data of states covid deaths/cases and daily tracking of them
ds = pd.read_csv('us.csv') #same data but no state specification (more general)
```

```python
df.head(10)
```

```python

df.rename(columns = {'date':'ObservationDate','state':'State','fips':'State Code','cases':'Cases Total','deaths':'Deaths Total'})
```

```python
df['date'] = pd.to_datetime(df['date'])


imputer = SimpleImputer(strategy='constant')
df2 = pd.DataFrame(imputer.fit_transform(df),columns=df.columns)

```

```python
states = df['state'].unique()
print(states) #total length is 56 because the dataset includes territories such as Guam, Puerto rico etc. as seen below
len(states)
```

```python
for idx in range(0,len(states)):
    C = df[df['state']==states[idx]].reset_index()        
    plt.scatter(np.arange(0,len(C)),C['cases'],color='white',label='Confirmed cases')
    
    plt.title(states[idx])
    plt.xlabel('Days since the first case(1/21/2020)')
    plt.ylabel('Number of cases ')
    plt.legend()
    plt.show() 
```

```python

for idx in range(0,len(states)):
    plt.style.use("dark_background")
    C = df[df['state']==states[idx]].reset_index()        
    plt.scatter(np.arange(0,len(C)),C['deaths'],color='white',label='Confirmed Deaths')
    
    plt.title(states[idx])
    plt.xlabel('Days since the first case(1/21/2020)')
    plt.ylabel('Number of cases ')
    plt.legend()
    plt.show() 
```
